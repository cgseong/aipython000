```python
# 간단한 사칙연산 프로그램 (변수 + 연산자 + 출력문만 사용)

a = 12  # 첫 번째 숫자
b = 5   # 두 번째 숫자

add_result = a + b   # 덧셈
sub_result = a - b   # 뺄셈
mul_result = a * b   # 곱셈
div_result = a / b   # 나눗셈(결과는 실수/float)

print("a =", a)  # a 출력
print("b =", b)  # b 출력

print("a + b =", add_result)  # 덧셈 결과 출력
print("a - b =", sub_result)  # 뺄셈 결과 출력
print("a * b =", mul_result)  # 곱셈 결과 출력
print("a / b =", div_result)  # 나눗셈 결과 출력
```
